```python
import pandas as pd
import numpy as np
from scipy import stats
from matplotlib import pyplot as plt
```

```python
s1 = np.random.normal(loc=0, scale=1, size=10)
s2 = np.random.normal(loc=0, scale=1, size=10)
```

```python
stats.ks_2samp(s1, s2)
```

```python
ks_res = []
res = []
res_a = []
res_w = []
res_t = []
for i in range(20):
    for size in range(10,1000, 10):
        #print(size)
        s1 = np.random.normal(loc=0, scale=1, size=size)
        s2 = np.random.normal(loc=0, scale=1, size=size)
        res.append((size, stats.ks_2samp(s1,s2)[1]))
        res_a.append((size, stats.anderson_ksamp([s1,s2])[2]))
        res_w.append((size, stats.wilcoxon(s1,s2)[1]))
        res_t.append((size, stats.ttest_ind(s1,s2)[1]))
```
