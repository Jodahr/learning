```python
import pandas as pd
import numpy as np
from scipy import stats
import pymc3 as pm
```

```python
np.random.seed(seed=42)
data = [stats.bernoulli(p=0.3).rvs(size) for size in range(100,1000,100)]
```

```python
#data
```

```python
with pm.Model() as model:
    p = pm.Uniform('p', 0, 1)   # prior
    likelihood = pm.Bernoulli("obs",p, observed=data[8]) # likelihood
```

```python
with model:
    trace = pm.sample(10000, step=[pm.Metropolis(), pm.NUTS()])  # sample from posterior
    pm.traceplot(trace, ["p"])  # plot histograms
```
